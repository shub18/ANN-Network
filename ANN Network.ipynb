```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import Sequential
```

```python
dataset=pd.read_csv('E:/Churn_Modelling.csv')
dataset
```

```python
x=dataset.iloc[:,3:13]
x
```

```python
y=dataset.iloc[:,-1]
y
```

```python
geography=pd.get_dummies(x['Geography'],drop_first=True)
gender=pd.get_dummies(x['Gender'],drop_first=True)
```

```python
X=pd.concat([x,geography,gender],axis=1)
X
```

```python
X=X.drop(['Geography','Gender'],axis=1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
model=ss.fit_transform(X)
```

```python
import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```python
se=Sequential()
se.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu',input_dim=11))
se.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu'))
se.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
se.compile(optimizer='Adamax',loss='binary_crossentropy',metrics=['accuracy'])
ann=se.fit(X_train, y_train, validation_split=0.33 ,batch_size=10, epochs=100)
```

```python
y_pred=se.predict(X_test)
y_pred=(y_pred>0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_pred,y_test)
ac
```
